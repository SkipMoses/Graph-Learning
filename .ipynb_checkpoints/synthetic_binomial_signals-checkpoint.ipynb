```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

def sigmoid(x):
  
    z = np.exp(-x)
    sig = 1 / (1 + z)

    return sig
def generate_signal(alpha, h, chi):
    P = (chi@h + alpha*np.ones(len(h)))
    X = [sigmoid(p) for p in P]
    for i in range(len(X)):
        test = np.random.uniform(0,1)
        if test < X[i]:
            X[i] = 1
        else:
            X[i] = 0
    return X
```

```python
G = nx.fast_gnp_random_graph(15, .2, seed = 123)
```

```python
nx.draw_circular(G)
```

```python
np.random.seed(123)
for s,t in G.edges:
    G[s][t]['weight'] = np.random.uniform(0,1)
```

```python
for s,t in G.edges:
    print(G.get_edge_data(s,t))
```

```python
edges,weights = zip(*nx.get_edge_attributes(G,'weight').items())
size = [w*5 for w in weights]
pos = nx.circular_layout(G)
ec = nx.draw_networkx_edges(G, pos, 
                            edgelist=edges, 
                            edge_color=weights, 
                            width=size,
                            edge_cmap=plt.cm.jet)
nc = nx.draw_networkx_nodes(G,pos, 
                            node_color='black', 
                            node_size = 100,
                            alpha = .9)
plt.colorbar(ec)
```

```python
L = nx.laplacian_matrix(G).toarray()
normL = (L.shape[0]/np.trace(L))*L
```

```python
w, chi = np.linalg.eig(normL)
```

```python
np.random.seed(123)
h = np.random.normal(0,1, 15)
np.random.seed(123)
alpha = np.random.uniform(0,1,100)
```

```python
Y = []
for a in alpha:
    Y.append(generate_signal(a,h,chi))
Y = np.transpose(np.array(Y))
Y.shape
```

```python
diffs = [np.transpose(i)@normL@i for i in np.transpose(Y)]
print(np.mean(diffs), np.std(diffs))
```

```python
# Creating histogram
fig, ax = plt.subplots(figsize =(10, 7))
ax.hist(diffs)
 
# Show plot
plt.show()
```

```python
np.random.seed(123)
sd = np.linalg.pinv(np.diag(w))
mu = np.zeros(15)
h2 = np.random.multivariate_normal(mu,sd)
alpha = np.random.uniform(0,1,100)
Y = []
for a in alpha:
    Y.append(generate_signal(a,h2,chi))
Y = np.transpose(np.array(Y))
```

```python
diffs = [np.transpose(i)@normL@i for i in np.transpose(Y)]
print(np.mean(diffs), np.std(diffs))
```

```python
# Creating histogram
fig, ax = plt.subplots(figsize =(10, 7))
ax.hist(diffs)
 
# Show plot
plt.show()
```
